# Small-Gap Experiment

```python
from pathlib import Path
import os
import sys


def _running_in_colab():
    try:
        import google.colab  # noqa: F401
        return True
    except ModuleNotFoundError:
        return False


def _mount_drive_if_needed():
    if not _running_in_colab():
        return
    if Path("/content/drive").exists():
        return
    try:
        from google.colab import drive
        drive.mount("/content/drive")
    except Exception as exc:
        print(f"Colab Drive mount skipped: {exc}")


def _walk_for_repo_root(base_dir, max_depth=4):
    base_dir = Path(base_dir)
    if not base_dir.exists():
        return None
    base_dir = base_dir.resolve()
    for root, dirs, _ in os.walk(base_dir):
        root_path = Path(root)
        depth = len(root_path.relative_to(base_dir).parts)
        if depth > max_depth:
            dirs[:] = []
            continue
        if "sp_bai" in dirs and (root_path / "sp_bai" / "__init__.py").exists():
            return root_path
    return None


def _find_repo_root():
    cwd = Path.cwd().resolve()
    direct_candidates = [cwd, *cwd.parents]
    direct_candidates += [
        Path("/content"),
        Path("/content/SP-BAI_AISTATS_Codes"),
        Path("/content/drive/MyDrive"),
        Path("/content/drive/MyDrive/SP-BAI_AISTATS_Codes"),
        Path("/content/drive/MyDrive/Colab Notebooks"),
        Path("/content/drive/MyDrive/Colab Notebooks/SP-BAI_AISTATS_Codes"),
    ]

    seen = set()
    for candidate in direct_candidates:
        candidate = Path(candidate)
        if not candidate.exists():
            continue
        candidate = candidate.resolve()
        if candidate in seen:
            continue
        seen.add(candidate)
        if (candidate / "sp_bai" / "__init__.py").exists():
            return candidate

    for search_base in [Path("/content"), Path("/content/drive/MyDrive")]:
        repo_root = _walk_for_repo_root(search_base, max_depth=5)
        if repo_root is not None:
            return repo_root
    return None


def ensure_repo_root_on_path():
    repo_root = _find_repo_root()
    if repo_root is None and _running_in_colab():
        _mount_drive_if_needed()
        repo_root = _find_repo_root()
    if repo_root is None:
        raise ModuleNotFoundError(
            "Could not locate the project root containing 'sp_bai'. "
            "Run the notebook from the project folder locally, or place the repo in Google Drive for Colab."
        )
    if str(repo_root) not in sys.path:
        sys.path.insert(0, str(repo_root))
    return repo_root


REPO_ROOT = ensure_repo_root_on_path()
print(f"Using repo root: {REPO_ROOT}")
```

```python
from sp_bai.experiments import SmallGapInstanceConfig, run_small_gap_instance

config = SmallGapInstanceConfig(
    dims=(10, 20),
    alpha=0.2,
    delta=1.0 / 10.0,
    R1=1 / 3,
    R2=1 / 3,
    n_simul=100,
    seed=100,
    lowrank_tol=1e-10,
    sbe_sample_cap=5_000_000,
    general_sample_cap=None,
    show_progress=True,
    print_every_sim=1,
    verbose=False,
)

run_small_gap_instance(config)
```
